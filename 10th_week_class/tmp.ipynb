```python
import csv
import matplotlib.pyplot as plt

f = open('/Users/inwon/Desktop/python_workspace/Python/10th_week_class/seoul.csv', encoding='euc-kr')
data = csv.reader(f)
next(data)
result = []

for row in data :
    if row[-1] != '' :
        if row[0].split('-')[1] == '03' and row[0].split('-')[2] == '10' :
            result.append(float(row[-1]))
#print(result)
plt.plot(result, 'hotpink')
plt.show()
```
