```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
from keras.optimizers import Adam
```

```python
data = pd.read_csv(r"/Users/inwon/Desktop/python_workspace/Python/11th_week_class/Iris.csv")

data = data.values

print(data)
```

```python
from sklearn.preprocessing import LabelEncoder

x = data[:, 0:4].astype(float)
y = data[:,-1]

encoder = LabelEncoder()
encoder.fit(y)
y_e = encoder.transform(y)
Y = np_utils.to_categorical(y_e)
```

```python
################
### 케라스 과제 ###
################


model = Sequential()

model.add(Dense(16, input_dim=4, activation='relu'))
model.add(Dense(3, activation='softmax'))

model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(x,Y,epochs=100, batch_size=1)

print(model.evaluate(x,Y)[1])
```

```python
##################
### 파이토치 과제 ###
##################

import torch
from torch import nn, optim
from torch.utils.data import DataLoader, Dataset
import torch.nn.functional as F
from sklearn.model_selection import train_test_split
```

```python
class Net(nn.Module):
    # define nn
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(4, 100)
        self.fc2 = nn.Linear(100, 100)
        self.fc3 = nn.Linear(100, 3)
        self.softmax = nn.Softmax(dim=1)

    def forward(self, X):
        X = F.relu(self.fc1(X))
        X = self.fc2(X)
        X = self.fc3(X)
        X = self.softmax(X)

        return X
```

```python
data = pd.read_csv(r"/Users/inwon/Desktop/python_workspace/Python/11th_week_class/Iris.csv")

data.loc[data.Species=='Iris-setosa','species']=0
data.loc[data.Species=='Iris-versicolor','species']=1
data.loc[data.Species=='Iris-virginica','species']=2

data = data.loc[:,['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','species']]

data
```

```python
train_x, test_x, train_y, test_y = train_test_split(data[data.columns[0:4]].values, data.species.values, test_size=0.8)
```

```python
from torch.autograd import Variable

train_x = Variable(torch.Tensor(train_x).float())
test_x = Variable(torch.Tensor(test_x).float())
train_y = Variable(torch.Tensor(train_y).long())
test_y = Variable(torch.Tensor(test_y).long())
```

```python
net = Net()

criterion = nn.CrossEntropyLoss()# cross entropy loss

optimizer = torch.optim.SGD(net.parameters(), lr=0.01)
```

```python
for epoch in range(1,1001):
    # optimizer.zero_grad()
    print("Epoch #", epoch)
    out = net(train_x)
    loss = criterion(out, train_y)
    print("The loss : ",loss.item())
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    
    # if epoch % 100 == 0:
    #     print ('number of epoch', epoch, 'loss', loss)
```

```python
pred = net(test_x)
```

```python
_, pred_y = torch.max(pred, 1)
```

```python
from sklearn.metrics import accuracy_score

print('Prediction acc: ',accuracy_score(test_y.data, pred_y.data))
```
